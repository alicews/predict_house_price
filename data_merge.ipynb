### Merge the sales data from different real estate companies and clean the data

```python
import pandas as pd
```

## Corcoran

```python
cor_sale = pd.read_csv("data/corcoran/sales.csv",encoding="unicode_escape")
print(cor_sale.shape)
```

```python
cor_df = cor_sale[["Title", "Apt", "Bedrooms", "Bathrooms", "Approx.Sq.Ft.", "Discription", "Image", "Price"]]
```

```python
cor_df.head(3)
```

```python
# remove the value such as "Corner of 6th Avenue and 2nd Street" since apartment number usually less than 7 characters
apt = []
for a in cor_df["Apt"]:
    a = a.replace("Apt. ","")
    if len(a) > 7:
        a = None
    apt.append(a)
```

```python
cor_df.loc[:, "Apt"] = apt
```

```python
cor_df.columns = ["Address", "Apt", "bedroom", "bathroom", "size", "intro", "photo", "price"]
cor_df.head(3)
```

# Douglas Elliman

```python
de_brooklyn_sale = pd.read_csv("data/Douglas_Elliman/Douglas_Elliman_brooklyn_sale_clean.csv",encoding="unicode_escape")
print("brooklyn", de_brooklyn_sale.shape)
de_bronx_sale = pd.read_csv("data/Douglas_Elliman/Douglas_Elliman_bronx_sale_clean.csv",encoding="unicode_escape")
print("bronx", de_bronx_sale.shape)
de_manhattan_sale = pd.read_csv("data/Douglas_Elliman/Douglas_Elliman_manhattan_sale_clean.csv",encoding="unicode_escape")
print("manhattan", de_manhattan_sale.shape)
de_queens_sale = pd.read_csv("data/Douglas_Elliman/Douglas_Elliman_queens_sale_clean.csv",encoding="unicode_escape")
print("queens", de_queens_sale.shape)
```

```python
df1 = de_brooklyn_sale[["house_address", "house_address3", "Beds", "Baths","Approximate Sq. Feet", "house_article", "image_url", "house_price"]]
df2 = de_bronx_sale[["house_address", "house_address3", "Beds", "Baths","Approximate Sq. Feet", "house_article", "image_url", "house_price"]]
df3 = de_manhattan_sale[["house_address", "house_address3", "Beds", "Baths","Approximate Sq. Feet", "house_article", "image_url", "house_price"]]
df4 = de_queens_sale[["house_address", "house_address3", "Beds", "Baths","Approximate Sq. Feet", "house_article", "image_url", "house_price"]]
```

```python
de_df = pd.concat([df1,df2,df3,df4])
de_df.columns = ["Address", "Apt", "bedroom", "bathroom", "size", "intro", "photo", "price"]
de_df.shape
```

```python
de_df.head(3)
```

```python
# clean house_address
building = []
apt = []
for name in de_df["Address"]:
    name = str(name).replace(" - ", ",").split(",")
#     print(name)
    if len(name) <=2:
        b = name[0]
        a = None
    else:        
        if name[0] == "Polhemus":
            b = name[0] + name[1]
            a = name[2].strip()
        if "St" in name[1]:
            b = name[1]
            a = name[2].strip()
        else: 
            b = name[0]
            if len(name[1]) <= 5:
                a = name[1].strip()
            else: 
                a = None
    building.append(b.strip())
    apt.append(a)
de_df["Address"] = building
de_df["Apt"] = apt
```

```python
de_df.head(3)
```

# Compass

```python
Compass_bronx_sale = pd.read_csv("data/Compass/Compass_bronx_compass_sale_done.csv",encoding="unicode_escape")
print("bronx", Compass_bronx_sale.shape)
Compass_brooklyn_sale = pd.read_csv("data/Compass/Compass_brooklyn_compass_sale_done.csv",encoding="unicode_escape")
print("brooklyn", Compass_brooklyn_sale.shape)
Compass_manhattan_sale = pd.read_csv("data/Compass/Compass_manhattan_compass_sale_done.csv",encoding="unicode_escape")
print("manhattan", Compass_manhattan_sale.shape)
Compass_queens_sale = pd.read_csv("data/Compass/Compass_queens_compass_sale_done.csv",encoding="unicode_escape")
print("queens", Compass_queens_sale.shape)
```

```python
df1 = Compass_bronx_sale[["Address", "Room", "bedroom", "bathroom", "size", "intro", "photo", "price"]]
df2 = Compass_brooklyn_sale[["Address", "Room", "bedroom", "bathroom", "size", "intro", "photo", "price"]]
df3 = Compass_manhattan_sale[["Address", "Apt", "bedroom", "bathroom", "size", "intro", "photo", "price"]]
df3.columns = ["Address", "Room", "bedroom", "bathroom", "size", "intro", "photo", "price"]
df4 = Compass_queens_sale[["Address", "Room", "bedroom", "bathroom", "size", "intro", "photo", "price"]]
```

```python
compass_df = pd.concat([df1,df2,df3,df4], sort=False)
compass_df.columns = ["Address", "Apt", "bedroom", "bathroom", "size", "intro", "photo", "price"]
```

```python
compass_df.head(3)
```

# merge all

```python
df = pd.concat([cor_df, de_df],sort=False)
s = []
for size in df["size"]:
    size = str(size)
    size = size.replace(",","")
    s.append(size)
p = []
for price in df["price"]:
    price = str(price)
    price = price.replace(",", "")
    price = price.replace("$", "")
    p.append(price)
df["size"] = s
df["price"] = p
```

```python
df_all = pd.concat([df, compass_df], sort=False)
```

```python
df_all = df_all.sort_values(by=['Address'])
```

# data cleaning

```python
df_all.head(3)
```

```python
df_all.loc[df_all["size"]=="nan","size"] = None
df_all.loc[df_all["Address"]=="nan","Address"] = None
```

```python
df_all = df_all[df_all["price"].isna()==False]
df_all = df_all[df_all["Address"].isna()==False]
df_all = df_all[df_all["Address"]!="null"]
df_all = df_all[(df_all["bedroom"].isna()==False)|(df_all["bathroom"].isna()==False)]
df_all = df_all[(df_all["Address"].isna()==False)]
df_all = df_all.iloc[:-5]
df_all = df_all.iloc[1:]
```

```python
df_all.isna().sum()/df_all.shape[0]
```

```python
# remove duplication
new = []
dup = []
df_all = df_all.sort_values(by=['Address', "Apt"])
for i in range(df_all.shape[0]-1): 
    if df_all.iloc[i][1] == df_all.iloc[i+1][1] and df_all.iloc[i][2] == df_all.iloc[i+1][2] and df_all.iloc[i][7] == df_all.iloc[i+1][7] and df_all.iloc[i][0] == df_all.iloc[i+1][0]:
        dup.append([df_all.iloc[i][0], df_all.iloc[i][1]])
    else:
        new.append(df_all.iloc[i].values)
```

```python
df_clean = pd.DataFrame(new)
df_clean.columns = ["Address", "Apt", "bedroom", "bathroom", "size", "intro", "photo", "price"]
```

```python
df_clean.head(3)
```

```python
bathroom = []
for i in df_clean['bathroom']:
    try:
        if i=="Half":
            bathroom.append(0.5)
        else:
            bathroom.append(float(i))
    except:
        bathroom.append(None)
price = []
for i in df_clean['price']:
    try:
        price.append(round(float(i) / 1000,0))
    except:
        price.append(None)
df_clean["bathroom"] = bathroom
df_clean["price(k)"] = price
```

```python
df_clean.shape
```

```python
df_clean.head(3)
```

```python
df_clean.loc[df_clean["bedroom"]==" ","bedroom"] = None
df_clean.loc[df_clean["size"]==" ","size"] = None
df_clean.loc[df_clean["size"]==" ","size"] = None
```

```python
df_clean.dtypes
```

```python
df_clean = df_clean[df_clean["bedroom"].isna()==False]
df_clean = df_clean[df_clean["bathroom"].isna()==False]
df_clean = df_clean[df_clean["size"].isna()==False]
print(df_clean.shape)
```

```python
df_clean.to_csv("data/all_sales.csv")
```
